# Merging tif with numpy array as return

```python
import rasterio
from rasterio.merge import merge
from rasterio.plot import show
import glob
import os
```

```python
# File and folder paths
dirpath = r"F:\India_elevation\Whole India 30 Arc Sec"
out_fp = r"F:\India_elevation\Whole India 30 Arc Sec\10n060e_20101117_gmted_dsc300.tif"
# Make a search criteria to select the DEM files
search_criteria = "*.tif"
q = os.path.join(dirpath, search_criteria)
print(q)
```

```python
dem_fps = glob.glob(q)
dem_fps
```

```python
src_files_to_mosaic = []
```

```python
for fp in dem_fps:
    src = rasterio.open(fp)
    src_files_to_mosaic.append(src)
src_files_to_mosaic
```

```python
# Merge function returns a single mosaic array and the transformation info
mosaic, out_trans = merge(src_files_to_mosaic)
```

```python
show(mosaic, cmap='terrain', vmin = 0, vmax = 1000)
```
